**Importing Pandas, Matplotlib, and Seaborn Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading and Displaying the First Few Rows of a CSV File of **Super Store** in Pandas DataFrame

```python
df = pd.read_csv('/content/SampleSuperstore.csv')
df.head()
```

Obtaining Information about the DataFrame

```python
df.info()
```

Generating Descriptive Statistics of a DataFrame

```python
df.describe()
```

Checking for missing values

```python
print(df.isnull().sum())
```

Creating a Histogram of Sales Distribution using Matplotlib and Seaborn

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=30)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

Identify correlations

```python
corr_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Visualize data: Profit by Category

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Profit', data=df)
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

Identify weak areas: Regions with low profit

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Profit', data=df)
plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

Visualize data: Profit by Sub-Category

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Sub-Category', y='Profit', data=df)
plt.title('Profit by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.xticks(rotation=90)
plt.show()
```

Visualize data: Sales by Region and Segment

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', hue='Segment', data=df)
plt.title('Sales by Region and Segment')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.legend(title='Segment')
plt.show()
```
